#### Задание 1. Соберите всю информацию о студентах в одну таблицу df. В получившейся таблице должна быть информация и оценки всех студентов из всех групп. Напечатайте несколько строк таблицы для демонстрации результата.¶

```python
table_count = 10

# создадим таблицу из общей информаци о студентах
frames1 = list()
for i in range(table_count):
    frames1.append(pd.read_csv("Data/Students_info_" + str(i) + ".csv", sep=","));
df1 = pd.concat(frames1);

# создадим таблицу из оценок всех студентов
frames2 = list()
for i in range(table_count):
    frames2.append(pd.read_csv("Data/Students_marks_" + str(i) + ".csv", sep=","));
df2 = pd.concat(frames2);

# соединим обе таблицы
df = pd.concat([df1, df2], axis=1)
```

**Задание 11. С помощью groupby выполните задания ниже. Также выведите время выполнения каждого из заданий.**

**Задание 12. Выполните задание 11 с помощью циклов. Сравните время выполнения.**

```python
# средний балл за экзамен по чтению среди каждой этнической группы
print("race/ethnicity groups; mean reading score:")
start_time = time.time()
race = set(df["race/ethnicity"])
for r in race:
    s, cnt = 0.0, 0.0
    for i in range(df.shape[0]):
        if df.iloc[i]["race/ethnicity"] == r:
            cnt += 1
            s += df.iloc[i]["reading score"]
    print(r + ":", s / cnt, sep=" ")
print("\nTime\n--- %s seconds ---\n" % (time.time() - start_time))
```

```python
# минимальный балл за экзамен по письму среди каждого уровня образования
print("education groups; minumum writing score:")
start_time = time.time()
education = set(df["education"])
for e in education:
    m = 1000
    for i in range(df.shape[0]):
        if df.iloc[i]["education"] == e:
            m = min(m, df.iloc[i]["writing score"])
    print(e + ":", m, sep=" ")
print("\nTime\n--- %s seconds ---\n" % (time.time() - start_time))
```

#### Задание 14. Сколько студентов успешно сдали экзамен по математике?

Создайте новый столбец в таблице df под названием Math_PassStatus и запишите в него F, если студент не сдал экзамен по математике (балл за экзамен < passmark), и P иначе.

Посчитайте количество студентов, сдавших и не сдавших экзамен по математике.

Сделайте аналогичные шаги для экзаменов по чтению и письму.

```python
df["Math_PassStatus"] = df["math score"].apply(lambda x: "F" if x < passmark else "P")
print("Pass/Fail in math:\n", df["Math_PassStatus"].value_counts(), '\n', sep="")

df["Reading_PassStatus"] = df["reading score"].apply(lambda x: "F" if x < passmark else "P")
print("Pass/Fail in reading:\n", df["Reading_PassStatus"].value_counts(), '\n', sep="")

df["Writing_PassStatus"] = df["writing score"].apply(lambda x: "F" if x < passmark else "P")
print("Pass/Fail in writing:\n", df["Writing_PassStatus"].value_counts(), '\n', sep="")

df
```

#### Задание 15. Сколько студентов успешно сдали все экзамены?

Создайте столбец OverAll_PassStatus и запишите в него для каждого студента 'F', если студент не сдал хотя бы один из трех экзаменов, а иначе 'P'.

Посчитайте количество студентов, которые сдали все экзамены.

```python
df["OverAll_PassStatus"] = df.apply(lambda x: "F" if (x["Math_PassStatus"] == "F") \
                                                      | (x["Reading_PassStatus"] == "F") \
                                                      | (x["Writing_PassStatus"] == "F") else "P", axis=1)
print("Students passed all the exams:", (df["OverAll_PassStatus"] == "P").sum())
df
```

#### Задание 16. Переведем баллы в оценки

### Система перевода баллов в оценки
####      90 и больше = A
####      80-89 = B
####      70-79 = C
####      60-69 = D
####      50-59 = E
####    меньше 50 = F (Fail)

Создайте вспомогательную функцию, которая будет по среднему баллу за три экзамена выставлять оценку студенту по данным выше критериям.

Создайте столбец Grade и запишите в него оценку каждого студента.

Выведите количество студентов, получивших каждую из оценок.

```python
def GetGrade(average_mark):
    if average_mark >= 90:
        return "A"
    if 80 <= average_mark <= 89:
        return "B"
    if 70 <= average_mark <= 79:
        return "C"
    if 60 <= average_mark <= 69:
        return "D"
    if 50 <= average_mark <= 59:
        return "E"
    return "F"

df["Grade"] = df.apply(lambda x: GetGrade((x["math score"] + x["reading score"] + x["writing score"]) / 3), axis=1)

print("Amount of students with grade:\n", df["Grade"].value_counts(), sep="")

df
```
